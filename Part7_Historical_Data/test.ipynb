```python
import requests
import defs
import pandas as pd
```

```python
session = requests.Session()
```

```python
instrument = "EUR_USD"
count = 10
granularity = "H1"
```

```python
url = f"{defs.OANDA_URL}/instruments/{instrument}/candles"
```

```python
url
```

```python
params = dict(
    count = count,
    granularity = granularity,
    price = "MBA"
)
```

```python
params
```

```python
response = session.get(url, params=params, headers=defs.SECURE_HEADER)
```

```python
response.status_code
```

```python
response.json()
```

```python
data = response.json()
```

```python
data.keys()
```

```python
len(data['candles'])
```

```python
for candle in data['candles']:
    print(candle)
```

```python
prices = ['mid', 'bid', 'ask']
ohlc = ['o', 'h', 'l', 'c']
```

```python
our_data = []
for candle in data['candles']:
    if candle['complete'] == False:
        continue
    new_dict = {
        'time': candle['time'],
        'volume': candle['volume']
    }
    for price in prices:
        for oh in ohlc:
            new_dict[f'{price}_{oh}'] = candle[price][oh]
    our_data.append(new_dict)
print(our_data[0])
```

```python
candles_df = pd.DataFrame.from_dict(our_data)
candles_df
```

```python
candles_df.to_pickle('EUR_USD_H1.pkl')
```
